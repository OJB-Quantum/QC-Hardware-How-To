```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Onri prefers using rcParams to increase the quality of all plots to be higher than the default settings.
# This only needs to be done once at the beginning of every Jupyter Notebook.
plt.rcParams['figure.dpi'] = 200

# Parameters for the wave
k = 2 * np.pi / 10  # Wave number (2*pi/wavelength)
omega = 2 * np.pi * 1  # Angular frequency (2*pi*frequency)
amplitude = 1  # Amplitude of the wave
length = 30  # Length of the wave
t = 0  # Time

# Generate coordinates for the wave
x = np.linspace(0, length, 500)
y = amplitude * np.sin(k * x - omega * t)  # Electric field component (E)
z = amplitude * np.cos(k * x - omega * t)  # Magnetic field component (B)

# Create a 3D plot
fig = plt.figure(figsize=(11, 7))
ax = fig.add_subplot(111, projection='3d')

# Turn off the 3D panes (the 'walls' behind the data)
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False

# Remove the pane edge lines as well
ax.xaxis.pane.set_edgecolor('none')
ax.yaxis.pane.set_edgecolor('none')
ax.zaxis.pane.set_edgecolor('none')

# Plot the electric and magnetic fields
ax.plot(x, y, np.zeros_like(x), label='Electric Field (E)', color='blue', linewidth=2)
ax.plot(x, np.zeros_like(x), z, label='Magnetic Field (B)', color='red', linewidth=2)

# Labels and legend
ax.set_xlabel('Propagation Direction (x)', labelpad=10)
# ax.set_ylabel('Electric Field Direction (y)', labelpad=10)
# ax.set_zlabel('Magnetic Field Direction (z)', labelpad=10)
ax.legend()

# Set the aspect ratio and remove grid
ax.set_box_aspect([2, 1, 1])  # Aspect ratio for better visualization
ax.grid(False)  # Remove grid

plt.title('Transverse Electromagnetic Wave')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Increase the quality of all plots using rcParams
plt.rcParams['figure.dpi'] = 200

# Parameters for the wave
k = 2 * np.pi / 10  # Wave number (2*pi/wavelength)
omega = 2 * np.pi * 1  # Angular frequency (2*pi*frequency)
amplitude = 1  # Amplitude of the wave
length = 30    # Length of the wave
t = 0          # Time

# Generate coordinates for the wave
x = np.linspace(0, length, 500)
y = amplitude * np.sin(k * x - omega * t)  # Electric field component (E)
z = amplitude * np.cos(k * x - omega * t)  # Magnetic field component (B)

# Create a 3D plot with a transparent background
fig = plt.figure(figsize=(10, 6))
# Make the figure facecolor transparent
fig.patch.set_alpha(0.0)

ax = fig.add_subplot(111, projection='3d')
# Make the Axes facecolor transparent
ax.patch.set_alpha(0.0)

# Turn off the 3D panes (the 'walls' behind the data)
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False

# Remove the pane edge lines as well
ax.xaxis.pane.set_edgecolor('none')
ax.yaxis.pane.set_edgecolor('none')
ax.zaxis.pane.set_edgecolor('none')

# Plot the electric and magnetic fields
ax.plot(x, y, np.zeros_like(x), label='Electric Field (E)', color='blue', linewidth=2)
ax.plot(x, np.zeros_like(x), z, label='Magnetic Field (B)', color='red', linewidth=2)

# Create polygons to simulate fill_between for Electric Field
for i in range(len(x) - 1):
    verts_e = [
        (x[i],     y[i],     0),
        (x[i + 1], y[i + 1], 0),
        (x[i + 1], 0,        0),
        (x[i],     0,        0),
    ]
    poly_e = Poly3DCollection([verts_e], color='blue', alpha=0.1)
    ax.add_collection3d(poly_e)

# Create polygons to simulate fill_between for Magnetic Field
for i in range(len(x) - 1):
    verts_b = [
        (x[i],     0, z[i]),
        (x[i + 1], 0, z[i + 1]),
        (x[i + 1], 0, 0),
        (x[i],     0, 0),
    ]
    poly_b = Poly3DCollection([verts_b], color='red', alpha=0.1)
    ax.add_collection3d(poly_b)

# Labels and legend
ax.set_xlabel('Propagation Direction (x)', labelpad=15)
ax.legend()

# Set the aspect ratio and remove grid
ax.set_box_aspect([2, 1, 1])  # Aspect ratio for better visualization
ax.grid(False)                # Remove grid

# Adjust the layout to prevent clipping
plt.tight_layout()

# Set the figure to save with a transparent background
# plt.title('Transverse Electromagnetic Wave with Shading')
plt.savefig('transparent_background_plot.png', transparent=True, dpi=300)
plt.show()
```
