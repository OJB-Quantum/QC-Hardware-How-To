### Display the Fraunhofer Pattern Equations

$$
\begin{aligned}
\Phi & =B w d \\
I_c(\Phi) & =I_c(0)\left|\frac{\sin \left(\pi \Phi / \Phi_0\right)}{\pi \Phi / \Phi_0}\right| \\
\Phi_0 & =\frac{h}{2 e}, \text { the magnetic flux quantum } \\
x & =\pi \Phi / \Phi_0 \rightarrow I_c(\Phi)=I_c(0)|\operatorname{sinc}(x)| \\
\operatorname{sinc}(x) & =\frac{\sin (x)}{x} \text { for } \mathrm{x} \neq 0
\end{aligned}
$$

---

Explanation of the Fraunhofer Pattern Equations
Magnetic Flux:

$$
\Phi=B w d
$$


Here, $B$ is the applied magnetic field, $w$ is the width of the junction, and $d$ is the effective thickness.
Critical Current:

$$
I_c(\Phi)=I_c(0)\left|\frac{\sin \left(\pi \Phi / \Phi_0\right)}{\pi \Phi / \Phi_0}\right|
$$


This equation describes the dependence of the critical current on the applied flux.
Magnetic Flux Quantum:

$$
\Phi_0=\frac{h}{2 e}
$$


This is the quantum of magnetic flux, where $h$ is Planck's constant and $e$ is the electron charge.
Dimensionless Argument:

$$
x=\pi \Phi / \Phi_0
$$


Substituting this into the critical current expression: $I_c(\Phi)=I_c(0)|\operatorname{sinc}(x)|$.
Sinc Function:

$$
\operatorname{sinc}(x)=\frac{\sin (x)}{x} \text { for } x \neq 0
$$


This is a normalized sinc function, often appearing in wave interference problems.

```python
import numpy as np
import matplotlib.pyplot as plt

# Onri prefers using rcParams to increase the quality of all plots to be higher than the default settings.
# This only needs to be done once at beginning of every Jupyter Notebook.
plt.rcParams['figure.dpi'] = 200

# High contrast plotting style via rcParams
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['figure.facecolor'] = '#000000'
plt.rcParams['axes.facecolor'] = '#000000'
plt.rcParams['axes.edgecolor'] = '#FFFFFF'
plt.rcParams['axes.labelcolor'] = '#FFFFFF'
plt.rcParams['xtick.color'] = '#FFFFFF'
plt.rcParams['ytick.color'] = '#FFFFFF'
plt.rcParams['text.color'] = '#FFFFFF'
plt.rcParams['font.size'] = 14
plt.rcParams['legend.fontsize'] = 14
plt.rcParams['lines.linewidth'] = 2.0
plt.rcParams['grid.color'] = '#444444'

# Uncomment if needed for inline plots
# %matplotlib inline

# # Optional: set a style that is available in your environment
# plt.style.use('ggplot')  # or comment this line out entirely

# Physical constants
Phi0 = 2.067833848e-15  # Magnetic flux quantum [Wb]
Ic0  = 1e-3             # Zero-flux critical current [A], e.g., 1 mA

# Define flux range in units of Phi0
flux_over_phi0 = np.linspace(-5, 5, 500)

# Fraunhofer current function
def fraunhofer_current(phi_over_phi0):
    # x = pi * Phi / Phi0
    x = np.pi * phi_over_phi0
    # Use np.sinc. Note that np.sinc(x) = sin(pi*x)/(pi*x).
    # We want sin(x)/x, so we set np.sinc(x/np.pi)
    return Ic0 * np.abs(np.sinc(x / np.pi))

# Compute the Fraunhofer pattern
Ic_vals = fraunhofer_current(flux_over_phi0)

# Plot
plt.figure(figsize=(7,5))
plt.plot(flux_over_phi0, Ic_vals, color='blue', label=r'$I_c(\Phi)$')
plt.xlabel(r'$\Phi / \Phi_0$')
plt.ylabel(r'$I_c(\Phi)$ (A)')
plt.title("Josephson Junction Fraunhofer Pattern")
plt.legend()
plt.show()
```
